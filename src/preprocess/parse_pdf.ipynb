```python
from parsestudio.parse import PDFParser

# Initialize with the desired parser backend
parser = PDFParser(parser="docling")  # Options: "docling", "pymupdf", "llama"
```

```python
root = "/storage/usmanova/judgeclaim/reports/original/"
# file = "AstraZeneca Sustainability Report 2023" #2022 Microsoft Environmental Sustainability Report

cur_ind = 0

print("PROCESSING: ", files[cur_ind])
print(root+files[cur_ind]+".pdf")
# outputs = parser.run([root+files[cur_ind]+".pdf"], modalities=["text", "tables"])
```

```python
file = "/storage/usmanova/reportkg/reports/original/Netflix ESG Report 2022.pdf"
outputs = parser.run(file, modalities=["text"])
```

```python
import os

out_path = "/storage/usmanova/judgeclaim/reports/processed/"
if not os.path.exists(out_path+files[cur_ind]):
    os.makedirs(out_path+files[cur_ind])

out_file = open(out_path+files[cur_ind]+"/content.txt", "w") 

for i in range(len(outputs)): 
    out_file.write(outputs[i].text.text) 
    out_file.write("\n") 
        
out_file.close() 
```

```python
import pandas as pd

def table2df(table):
    lines = table.split("\n")
    header = lines[1].strip("|").split("|")

    data = [] 

    # Loop through lines starting from 2
    for line in lines[2:]:
        
        # Break once we hit an empty line
        if not line.strip():
            break
            
        cols = line.strip("|").split("|")
        row = dict(zip(header, cols))
        data.append(row)
    
    df = pd.DataFrame(data)
    return df
```

```python
import pandas as pd

t = 0
for table in outputs[0].tables:
    # print(t)
    df = table2df(table.markdown)
    df.to_csv(out_path+files[cur_ind]+f"/table_{t}.csv")
    t += 1
```
